```python
import json

# Load the JSON file
with open('SC_01_HypoF_thermout.json', 'r') as f:
    data = json.load(f)

# Print the top-level keys
print("Top-level keys:")
for key in data.keys():
    print(key)

# Explore the structure of the data
def explore_data(data, indent=0):
    if isinstance(data, dict):
        for key, value in data.items():
            print(' ' * indent + str(key))
            explore_data(value, indent + 2)
    elif isinstance(data, list):
        for item in data:
            explore_data(item, indent + 2)
    else:
        print(' ' * indent + str(data))

print("\nData structure:")
explore_data(data)
```

```python
# Approach 1: Using a context manager to write directly to the file
with open('data_structure.txt', 'w') as f:
    # Print top-level keys
    f.write("Top-level keys:\n")
    for key in data.keys():
        f.write(f"{key}\n")
    
    # Print full structure
    f.write("\nData structure:\n")
    def explore_data(data, indent=0):
        if isinstance(data, dict):
            for key, value in data.items():
                f.write(' ' * indent + str(key) + '\n')
                explore_data(value, indent + 2)
        elif isinstance(data, list):
            for item in data:
                explore_data(item, indent + 2)
        else:
            f.write(' ' * indent + str(data) + '\n')
    
    explore_data(data)
```
